```python
import joblib
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Load the saved DataFrame
df_prices = joblib.load(r"C:\Users\FEMI\Desktop\token_correlation_heatmap\Database\df_prices.joblib")

print(df_prices.head())
```

## token correlation heatmap


```python
# Calculate the correlation matrix
corr = df_prices.corr()

# Plot the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title("Token Correlation Heatmap")
plt.show()
```

```python
# Compute volatility (standard deviation of returns) for each token

# Calculate log returns
returns = df_prices.pct_change().apply(lambda x: np.log1p(x))

# Compute volatility (standard deviation of log returns)
volatility = returns.std()

print("Volatility for each token:")
print(volatility)
```
